```python
#Pra pemrosesan korelasi pearson
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Membaca dataset
df = pd.read_csv('IRIS.csv')
#Menghitung matriks korelasi pearson
corr_matrix = df.corr(method='pearson')
#Membuat heatmap dari matriks korelasi
sns.heatmap(corr_matrix, annot=True)
#Menampilkan grafik
plt.show()
```

```python
#3.Separate the feature and target variable from the dataframe
x = df.iloc[:, :-1]#Features
y = df.iloc[:, -1]#Target

print(y)
```

```python
#4.Create a Decision Tree Classifier object
dtc = DecisionTreeClassifier()
```

```python
#5.Fit the classfier to the data using the fit() method
dtc.fit(x, y)
```

```python
#6.Visualize the decision tree using the plot_tree() function from
plt.figure(figsize=(12,8))
plot_tree(dtc, feature_names=x.columns, class_names=y.unique(), filled=True)
plt.show()
```

```python
#split training testing datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
RANDOM_SEED = 0
dt_clf = DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=RANDOM_SEED)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
dt_clf.fit(x_train, y_train)
dt_clf.score(x_train,y_train),dt_clf.score (x_test, y_test)
```

```python
#Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
predictions = dt_clf.predict(x_test)
conf_matrix = confusion_matrix(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
conf_matrix,accuracy
```

```python
print('Decision Tree Data Mining Model', classification_report(y_test,predictions))
```
